## Galileo's Leaning Tower of Pisa experiment

https://en.wikipedia.org/wiki/Galileo%27s_Leaning_Tower_of_Pisa_experiment

Between 1589â€“92,the Italian scientist Galileo Galilei (then professor of mathematics at the University of Pisa) is said to have dropped two spheres of different masses from the Leaning Tower of Pisa to demonstrate that their time of descent was independent of their mass, according to a biography by Galileo's pupil Vincenzo Viviani, composed in 1654 and published in 1717

```r
set.seed(1)

g <- 9.81
n <- 25
# time in sequence
x <- rnorm(n, sd = 1)
# seq(start, end, length)
timeSeq <- seq(0, 3.4, len = n)
```

```r
x
```

```r
timeSeq
```

```r
y = 56.67 - 0.5 * g * timeSeq^2 + x
y
```

Construct a Data Frame

```r
pisa.df <- as.data.frame(cbind(timeSeq, y))
head(pisa.df)
```

### Plot

```r
library(ggplot2)
```

```r
ggplot(pisa.df, aes(x = timeSeq, y = y)) + geom_point()
```
